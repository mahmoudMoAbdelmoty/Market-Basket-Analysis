### `Data Preparation for Market Analysis`

```python
df.columns
```

```python
df['Single Transaction'] = df['Member_number'].astype(str) + '_' + df['Date'].astype(str)
```

```python
df.head()
```

```python
df2 = pd.crosstab(df['Single Transaction'], df['itemDescription'])
```

```python
df2.head()
```

```python
basket_input = df2.applymap(lambda x: 1 if x>0 else 0 )
```

```python
basket_input.sample(10)
```

### `Build Apriori Alg`

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
frequent_pattern = apriori(basket_input, min_support=0.001, use_colnames=True)
rules = association_rules(frequent_pattern, metric='lift')
```

```python
frequent_pattern.sample(15)
```

```python
rules.head()
```

- `antecedents` and `consequents` columns show items that are frequently purchased together.

```python
# to get the most frequent item combination
rules.sort_values(by=['support', 'confidence', 'lift'], axis=0, ascending=False).head(10)
```

table shows that the four most popular product combinations that are frequently bought together are: 
- Rolls and milk
- Yogurt and milk
- Sausages and milk
- Soda and vegetables
